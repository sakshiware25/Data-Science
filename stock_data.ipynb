```python
#to convert that usd,dollar,$ into usd as they are all same ..so will use convertor
def standardize_currency(curr):
    if curr == "$$" or curr == "Dollar":
        return"USD"
    return curr



df_financials = pd.read_excel(r"C:\Users\Ram Gadekar\Downloads\chapter3_assets\chapter_3_assets\3_read_write_to_excel\movies_db.xlsx","financials",converters={
    'currency': standardize_currency})
df_financials.head(5)
```

```python
#to merge financial and movies 

df_merged =pd.merge(df_movies,df_financials,on="movie_id")
df_merged.head(4)
```

```python
df_merged.to_excel("movies_merged.xlsx",sheet_name="merged",index= False)
```
